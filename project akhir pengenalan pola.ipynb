```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from keras.models import load_model
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, Flatten
from keras.models import Model
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns

def LoadCitraTraining(sDir, LabelKelas):
    JumlahKelas = len(LabelKelas)
    TargetKelas = np.eye(JumlahKelas)

    X = []  # Menampung Data Citra
    T = []  # Menampung Target

    print("=== LOADING DATA ===")
    total_data = 0

    for i in range(JumlahKelas):
        DirKelas = os.path.join(sDir, LabelKelas[i])
        files = os.listdir(DirKelas)
        class_count = 0

        print(f"Memuat kelas '{LabelKelas[i]}':")
        for f in files:
            ff = f.lower()
            if ff.endswith('.jpg') or ff.endswith('.jpeg') or ff.endswith('.png'):
                NmFile = os.path.join(DirKelas, f)
                # Perbaikan: Konversi BGR ke RGB untuk warna yang benar
                img = cv2.imread(NmFile, 1)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Konversi BGR ke RGB
                img = cv2.resize(img, (128, 128))
                img = np.asarray(img) / 255.0
                img = img.astype('float32')
                X.append(img)
                T.append(TargetKelas[i])
                class_count += 1

        print(f"  - Jumlah data: {class_count}")
        total_data += class_count

    X = np.array(X)
    T = np.array(T)
    X = X.astype('float32')
    T = T.astype('float32')

    print(f"\nTotal data keseluruhan: {total_data}")
    print(f"Jumlah kelas: {JumlahKelas}")
    print(f"Shape data X: {X.shape}")
    print(f"Shape data T: {T.shape}")

    return X, T

def ModelDeepLearningCNN(JumlahKelas):
    input_img = Input(shape=(128, 128, 3))
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
    x = MaxPooling2D((2, 2), padding='same')(x)
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2), padding='same')(x)
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    x = Flatten()(x)
    x = Dense(100, activation='relu')(x)
    x = Dense(100, activation='relu')(x)
    x = Dense(JumlahKelas, activation='softmax')(x)
    ModelCNN = Model(input_img, x)
    ModelCNN.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return ModelCNN

def TrainingCNN(JumlahEpoh, DirektoriDataSet, LabelKelas, NamaFileBobot='weight.h5'):
    # Load semua data
    X, D = LoadCitraTraining(DirektoriDataSet, LabelKelas)
    JumlahKelas = len(LabelKelas)

    # PERBAIKAN: Split data menjadi training dan testing
    print("\n=== PEMBAGIAN DATA ===")
    X_train, X_test, D_train, D_test = train_test_split(
        X, D, test_size=0.2, random_state=42,
        stratify=np.argmax(D, axis=1)  # Stratified split untuk distribusi seimbang
    )

    print(f"Data Training: {len(X_train)} ({len(X_train)/len(X)*100:.1f}%)")
    print(f"Data Testing: {len(X_test)} ({len(X_test)/len(X)*100:.1f}%)")

    # Tampilkan distribusi kelas untuk training dan testing
    print("\nDistribusi kelas pada data training:")
    for i, label in enumerate(LabelKelas):
        count = np.sum(np.argmax(D_train, axis=1) == i)
        print(f"  {label}: {count} data")

    print("\nDistribusi kelas pada data testing:")
    for i, label in enumerate(LabelKelas):
        count = np.sum(np.argmax(D_test, axis=1) == i)
        print(f"  {label}: {count} data")

    # Buat dan training model
    ModelCNN = ModelDeepLearningCNN(JumlahKelas)
    print(f"\n=== MULAI TRAINING ===")
    print("Arsitektur Model:")
    ModelCNN.summary()

    # Training dengan validation split dari data training
    history = ModelCNN.fit(
        X_train, D_train,
        epochs=JumlahEpoh,
        shuffle=True,
        validation_split=0.2,  # 20% dari data training untuk validasi
        verbose=1
    )

    # Simpan model
    ModelCNN.save(NamaFileBobot)
    print(f"\nModel disimpan sebagai: {NamaFileBobot}")

    return ModelCNN, history, X_test, D_test

def EvaluateModel(ModelCNN, X_test, D_test, LabelKelas):
    """
    PERBAIKAN: Fungsi untuk evaluasi model pada data testing
    """
    print("\n=== EVALUASI MODEL PADA DATA TESTING ===")

    # Prediksi pada data testing
    predictions = ModelCNN.predict(X_test)

    # Konversi one-hot ke label
    y_true = np.argmax(D_test, axis=1)
    y_pred = np.argmax(predictions, axis=1)

    # Hitung akurasi
    test_accuracy = accuracy_score(y_true, y_pred)
    print(f"Akurasi pada data testing: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")

    # Classification report
    print(f"\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=LabelKelas))

    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=LabelKelas, yticklabels=LabelKelas)
    plt.title('Confusion Matrix - Data Testing')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.show()

    return test_accuracy, predictions

def ShowSamplePredictions(ModelCNN, X_test, D_test, LabelKelas, num_samples=6):
    """
    PERBAIKAN: Tampilkan contoh hasil prediksi
    """
    predictions = ModelCNN.predict(X_test)

    # Pilih sample secara random
    indices = np.random.choice(len(X_test), num_samples, replace=False)

    plt.figure(figsize=(15, 10))
    for i, idx in enumerate(indices):
        plt.subplot(2, 3, i+1)

        # Tampilkan gambar (sudah dalam format RGB)
        plt.imshow(X_test[idx])

        # Label asli dan prediksi
        true_label = LabelKelas[np.argmax(D_test[idx])]
        pred_label = LabelKelas[np.argmax(predictions[idx])]
        confidence = np.max(predictions[idx])

        # Warna hijau jika benar, merah jika salah
        color = 'green' if true_label == pred_label else 'red'
        plt.title(f'True: {true_label}\nPred: {pred_label}\nConf: {confidence:.3f}',
                 color=color, fontsize=10)
        plt.axis('off')

    plt.suptitle('Contoh Hasil Prediksi pada Data Testing', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

def Klasifikasi(DirDataSet, DirKlasifikasi, LabelKelas, ModelCNN=None):
    if ModelCNN is None:
        ModelCNN = load_model('weight.h5')

    X = []
    ls = []
    DirKelas = os.path.join(DirDataSet, DirKlasifikasi)
    print("Membaca file dari:", DirKelas)
    files = os.listdir(DirKelas)

    for f in files:
        ff = f.lower()
        print(f)
        if ff.endswith('.jpg') or ff.endswith('.jpeg') or ff.endswith('.png'):
            ls.append(ff)
            NmFile = os.path.join(DirKelas, f)
            img = cv2.imread(NmFile, 1)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Perbaikan warna
            img = cv2.resize(img, (128, 128))
            img = np.asarray(img) / 255.0
            img = img.astype('float32')
            X.append(img)

    X = np.array(X)
    X = X.astype('float32')
    hs = ModelCNN.predict(X)

    LKlasifikasi = []
    LKelasCitra = []

    for i in range(X.shape[0]):
        v = hs[i, :]
        if v.max() > 0.5:
            idx = int(np.argmax(v))
            LKelasCitra.append(LabelKelas[idx])
        else:
            idx = -1
            LKelasCitra.append("-")
        LKlasifikasi.append(idx)

    LKlasifikasi = np.array(LKlasifikasi)
    return ls, hs, LKelasCitra

# ===== PROGRAM UTAMA =====

# Pastikan path ini sesuai dengan lokasi folder hasil ekstrak dataset kamu
DirektoriDataSet = "/content/content/dataset_gambar"

# Nama folder kelas harus sesuai dengan folder di dalam dataset
LabelKelas = ["banana unripe", "banana semi-ripe", "banana fully-ripe"]

JumlahEpoh = 20
FileBobot = "WRambu.h5"

print("=" * 60)
print("SISTEM KLASIFIKASI KEMATANGAN PISANG")
print("=" * 60)

# PERBAIKAN: Training model CNN dengan train-test split
ModelCNN, history, X_test, D_test = TrainingCNN(JumlahEpoh, DirektoriDataSet, LabelKelas, FileBobot)

print("\n" + "="*50)
print("ARSITEKTUR MODEL:")
print("="*50)
ModelCNN.summary()

# PERBAIKAN: Plot hasil training (loss dan accuracy) dengan validation
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# PERBAIKAN: Evaluasi model pada data testing
test_accuracy, test_predictions = EvaluateModel(ModelCNN, X_test, D_test, LabelKelas)

# PERBAIKAN: Tampilkan contoh prediksi
ShowSamplePredictions(ModelCNN, X_test, D_test, LabelKelas)

# PERBAIKAN: Ringkasan hasil
print("\n" + "="*50)
print("RINGKASAN HASIL AKHIR")
print("="*50)
print(f"Akurasi Training Terakhir: {history.history['accuracy'][-1]:.4f} ({history.history['accuracy'][-1]*100:.2f}%)")
print(f"Akurasi Validation Terakhir: {history.history['val_accuracy'][-1]:.4f} ({history.history['val_accuracy'][-1]*100:.2f}%)")
print(f"Akurasi Testing: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f"Model disimpan sebagai: {FileBobot}")
print(f"Total epoch training: {JumlahEpoh}")
print("="*50)
```
